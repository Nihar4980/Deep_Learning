#### Checking stationary of data

```python
def adf_test(series):
    result=adfuller(series)
    print('ADF Statistics: {}'.format(result[0]))
    print('p- value: {}'.format(result[1]))
    if result[1] <= 0.05:
        print("strong evidence against the null hypothesis, reject the null hypothesis. Data has no unit root and is stationary")
    else:
        print("weak evidence against null hypothesis, time series has a unit root, indicating it is non-stationary ")
```

## split data into train test

```python
train_dataset_end = datetime(2018,12,31)
timedelta(days=1)
```

```python
train_data = data_multi[:train_dataset_end]
test_data = data_multi[train_dataset_end+timedelta(days=1):]
```

```python
train_data
```

```python
train_data
```

```python
print(train_data.shape, test_data.shape)
```

```python
scaler = MinMaxScaler()
scaler.fit(train_data)
scaler_train = scaler.transform(train_data)
scaler_test = scaler.transform(test_data)
```

```python
scaler_train
```

```python
scaler_train[:1,:].shape
```

```python
scaler_test
```

```python
print(scaler_train.shape, scaler_test.shape)
```

```python
trainX = []
trainY = []

n_future = 1   # Number of days we want to look into the future based on the past days.
n_past = 14  # Number of past days we want to use to predict the future.

#Reformat input data into a shape: (n_samples x timesteps x n_features)
#In my example, my scaler_train has a shape (96431, 4)
#96431 refers to the number of data points and 4 refers to the columns (multi-variables).
for i in range(n_past, len(scaler_train) - n_future +1):
    trainX.append(scaler_train[i - n_past:i, 0:train_data.shape[1]])
    trainY.append(scaler_train[i + n_future - 1:i + n_future, 0])

X_train, y_train = np.array(trainX), np.array(trainY)

print('X_train shape == {}'.format(X_train.shape))
print('y_train shape == {}'.format(y_train.shape))
```

```python
train_data_temp = train_data.iloc[14:,:1]
train_data_temp
```

```python
train_data_temp.shape
```

```python
testX = []
testY = []

n_future = 1   # Number of days we want to look into the future based on the past days.
n_past = 14  # Number of past days we want to use to predict the future.

#Reformat input data into a shape: (n_samples x timesteps x n_features)
#In my example, my scaler_train has a shape (96431, 4)
#96431 refers to the number of data points and 4 refers to the columns (multi-variables).
for i in range(n_past, len(scaler_test) - n_future +1):
    testX.append(scaler_test[i - n_past:i, 0:test_data.shape[1]])
    testY.append(scaler_test[i + n_future - 1:i + n_future, 0])

X_test, y_test = np.array(testX), np.array(testY)

print('X_train shape == {}'.format(X_test.shape))
print('y_train shape == {}'.format(y_test.shape))
```

```python
test_data_temp = test_data.iloc[14:,:1]
test_data_temp
```

```python
test_data_temp.shape
```

```python
callback = EarlyStopping(
    monitor="val_loss",
    min_delta=0.00001,
    patience=4,
    verbose=1,
    mode="auto",
    baseline=None,
    restore_best_weights=False
)
```

```python
model_gru = Sequential()
model_gru.add(GRU(128, return_sequences=True, input_shape= ((X_train.shape[1], X_train.shape[2])), activation='relu'))
model_gru.add(GRU(units=64, return_sequences=True,activation='relu'))
model_gru.add(GRU(units=32, activation='relu'))

model_gru.add(Dense(units=16, activation='relu'))
model_gru.add(Dropout(0.2))
model_gru.add(Dense(y_train.shape[1], activation='linear'))
model_gru.compile(loss='mse', optimizer='adam')
```

```python
model_gru.summary()
```

```python
model_gru.get_weights()
```

```python
# fit network
gru_history = model_gru.fit(X_train, y_train, epochs=10, batch_size=32, validation_split=0.1, verbose=1, callbacks=callback)
```

```python
plt.plot(gru_history.history['loss'], label='loss', color='brown')
plt.plot(gru_history.history['val_loss'], label='val loss',c='c')
plt.legend()
plt.show()
```

```python
pred_train_gru = model_gru.predict(X_train)
```

```python
pred_train_gru
```

```python
prediction_copies = np.repeat(pred_train_gru, train_data.shape[1], axis=-1)
temp_gru_pred_train = scaler.inverse_transform(prediction_copies)[:,0]

temp_gru_pred_train.shape
```

```python
train_data_temp['temp_gru'] = temp_gru_pred_train
```

```python
pred_test_gru = model_gru.predict(X_test)
pred_test_gru 
```

```python
prediction_copies = np.repeat(pred_test_gru, test_data.shape[1], axis=-1)
temp_gru_pred_test = scaler.inverse_transform(prediction_copies)[:,0]

temp_gru_pred_test.shape
```
