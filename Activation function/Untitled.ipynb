```python
import numpy as np
import matplotlib.pyplot as plt

# Linear activation function and its derivative
def linear_activation(x, w=2, b=1):
    return w * x + b

def derivative_linear_activation(x, w=2):
    return w  # Derivative of a linear function is a constant (the weight)

# Generate x values
x_values = np.linspace(-10, 10, 100)

# Calculate y values for the linear activation function and its derivative
y_values_linear_activation = linear_activation(x_values)
y_values_derivative_linear_activation = derivative_linear_activation(x_values)

# Plot the linear activation function
plt.plot(x_values, y_values_linear_activation, label='Linear Activation: f(x) = 2x + 1', color='blue')

# Plot the derivative of the linear activation function
plt.plot(x_values, y_values_derivative_linear_activation, label="Derivative: f'(x) = 2 (constant)", linestyle='--', color='red')

# Set labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Activation Function and Its Derivative')

# Add a legend
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def softmax(x):
    exp_x = np.exp(x - np.max(x))  # Subtracting the maximum value for numerical stability
    return exp_x / exp_x.sum(axis=0)

# Generate a range of logits
logits_range = np.linspace(-5, 5, 100)

# Compute softmax probabilities for each set of logits
softmax_probs = np.array([softmax([logit, 1.0, 0.1]) for logit in logits_range]).T

# Plot the softmax probabilities
plt.figure(figsize=(8, 6))
plt.plot(logits_range, softmax_probs[0], label='Class 1', color='blue')
plt.plot(logits_range, softmax_probs[1], label='Class 2', color='green')
plt.plot(logits_range, softmax_probs[2], label='Class 3', color='red')

plt.xlabel('Logits')
plt.ylabel('Softmax Probabilities')
plt.title('Softmax Function')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def softmax(x):
    exp_x = np.exp(x - np.max(x))
    return exp_x / exp_x.sum(axis=0)

def softmax_derivative(logit_range, class_index):
    softmax_probs = np.array([softmax([logit, 1.0, 0.1]) for logit in logit_range]).T
    return softmax_probs[class_index] * (np.eye(len(logit_range)) - softmax_probs[class_index][:, np.newaxis])

# Generate a range of logits
logits_range = np.linspace(-5, 5, 100)

# Choose a specific class index for the derivative
class_index = 0

# Compute the derivative of the softmax function for the chosen class
softmax_derivative_values = softmax_derivative(logits_range, class_index)

# Plot the derivative of the softmax function
plt.figure(figsize=(8, 6))
for i in range(len(logits_range)):
    plt.plot(logits_range, softmax_derivative_values[i])

plt.xlabel('Logits')
plt.ylabel(f'Derivative of Softmax (Class {class_index+1})')
plt.title('Derivative of Softmax Function')
plt.legend()
plt.grid(True)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(9, 5))
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
#plt.ylim(-6,6)
for i in range(len(logits_range)):
    ax.plot(logits_range, softmax_derivative_values[i])
    
plt.savefig("soft_max_der.png") 
fig.show()
```
