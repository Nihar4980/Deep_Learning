```python
import keras
from keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
```

```python
img = keras.utils.load_img('data/cat.jpg',target_size=(200,200))
```

```python
plt.imshow(img)
```

```python
type(img)
```

```python
img = keras.utils.img_to_array(img)
```

```python
img
```

```python
img.shape
```

```python
input_batch = img.reshape(1,200,200,3)
input_batch.shape
```

https://blog.keras.io/building-powerful-image-classification-models-using-very-little-data.html

```python
# nearest_fill
datagen = ImageDataGenerator(
        rotation_range=30,
        width_shift_range=0.2,
        height_shift_range=0.2,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        fill_mode='nearest'
        )
```

```python
i = 0
for batch in datagen.flow(input_batch, batch_size=1,save_to_dir='data/aug/aug_nearest/'):
    i += 1
    if i > 10:
        break
```

```python
# constant_fill
datagen = ImageDataGenerator(
        rotation_range=30,
        width_shift_range=0.2,
        height_shift_range=0.2,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        fill_mode='constant'
    )
```

```python
i = 0
for batch in datagen.flow(input_batch, batch_size=1,save_to_dir='data/aug/aug_constant//'):
    i += 1
    if i > 10:
        break
```

```python
# reflect_fill
datagen = ImageDataGenerator(
        rotation_range=30,
        width_shift_range=0.2,
        height_shift_range=0.2,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        fill_mode='reflect'
    )
```

```python
i = 0
for batch in datagen.flow(input_batch, batch_size=1,save_to_dir='data/aug/aug_reflect/'):
    i += 1
    if i > 10:
        break
```
