```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('DailyDelhiClimateTest.csv')
df = df[['date','meantemp']]
df.head()
```

```python
df.shape
```

```python
plt.scatter(df['date'],df['meantemp'],c='r')
```

```python
df['ewma1'] = df['meantemp'].ewm(alpha=0.9).mean()
```

```python
df['ewma2'] = df['meantemp'].ewm(alpha=0.75).mean()
```

```python
df['ewma3'] = df['meantemp'].ewm(alpha=0.5).mean()
```

```python
df['ewma4'] = df['meantemp'].ewm(alpha=0.25).mean()
```

```python
df['ewma5'] = df['meantemp'].ewm(alpha=0.1).mean()
```

```python
df
```

```python
plt.scatter(df['date'],df['meantemp'],c='r')
plt.plot(df['date'],df['ewma1'],c='k')
plt.plot(df['date'],df['ewma2'],c='g')
plt.plot(df['date'],df['ewma3'],c='b')
plt.plot(df['date'],df['ewma4'],c='c')
plt.plot(df['date'],df['ewma5'],c='pink')
plt.show()
```

```python
plt.scatter(df['date'],df['meantemp'],c='r')
plt.plot(df['date'],df['ewma1'],c='k')
plt.show()
```
