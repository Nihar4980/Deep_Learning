```python
from keras.models import Model
```

```python
from keras.layers import *
```

```python
# One Input with two output prediction
```

```python
x = Input(shape=(3,))
```

```python
hidden1 = Dense(128,activation='relu')(x)
hidden2 = Dense(64,activation='relu')(hidden1)

output1 = Dense(1,activation='linear')(hidden2)

output2 = Dense(1,activation='sigmoid')(hidden2)
```

```python
model = Model(inputs = x, outputs= [output1, output2])
```

```python
model.summary()
```

```python
from keras.utils import plot_model
plot_model(model)
```

```python
plot_model(model,show_shapes=1)
```

```python
# Two input with One output prediction
```

```python
# Define two sets of inputs
input1 = Input(shape=(32,))
input2 = Input(shape=(128,))
```

```python
# first branch operate with first input
x = Dense(128,activation='relu')(input1)
x1 = Dense(64,activation='relu')(x)
```

```python
# second branch operate with second input
y = Dense(128,activation='relu')(input2)
y1 = Dense(64,activation='relu')(y)
y2 = Dense(32,activation='relu')(y1)
```

```python
# combined the output of two branches

combined = concatenate([x1, y2])
```

```python
# apply a FC layer and then a regression prediction on the
# combined outputs
z = Dense(16, activation="relu")(combined)
z1 = Dense(1, activation="linear")(z)
```

```python
# our model will accept the inputs of the two branches and
# then output a single value
model = Model(inputs=[input1, input2], outputs=z1)
```

```python
plot_model(model)
```

```python
plot_model(model, show_shapes=1)
```
